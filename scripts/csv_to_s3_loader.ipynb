```python
from datetime import datetime
import boto3

s3 = boto3.client("s3")

today = datetime.now()

year = today.strftime("%Y")
month = today.strftime("%m")
day = today.strftime("%d")

local_file = f"transactions_{year}_{month}_{day}.csv"

s3_key = f"raw/{year}/{month}/{day}/transactions.csv"

s3.upload_file(
    local_file,
    "fintech-payments",
    s3_key
)

print(f"Uploaded to s3://fintech-payment-monitoring/{s3_key}")
```
